```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans 
```

```python
x = df.iloc[:, [3,4 ]].values  
```

```python
x
```

```python
x[:,0] 
```

```python
x[:,1] 
```

```python
plt.scatter(x[:,0],x[:,1])
#plot
```

```python
wcss_list= []   
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elbow Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  

#Elbow Method Graph
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x)

#kmeans model used
```

```python
print(y_predict) #Predicted labels
```

```python
df['cluster']=y_predict #adding new column named cluster to dataset which contains predicted labels for data points
print(df)
```

```python
print(kmeans.cluster_centers_)
#cluster centres
```

```python
plt.scatter(x[:,0],x[:,1],c=y_predict,cmap='rainbow')
plt.plot(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],'.',color='black')
plt.title('Distribution of Questions for 2024')  
plt.xlabel('Answer')  
plt.ylabel('Question')  
plt.show()  
#Based on the historical trends, develop a predictive model to forecast the potential distribution of questions for the 2024 exam
#The distribution of questions for 2024 are across all the different subjects. 
#clusters used for distribution of questions for 2024
```

```python
plt.figure(figsize = (20,10))
plt.scatter(x[y_predict == 0,0],x[y_predict == 0,1],s = 50, c = 'green', label = "0")
plt.scatter(x[y_predict == 1,0],x[y_predict == 1,1],s = 50, c = 'blue', label = "1")
plt.scatter(x[y_predict == 2,0],x[y_predict == 2,1],s = 50, c = 'yellow', label = "2")
plt.scatter(x[y_predict == 3,0],x[y_predict == 3,1],s = 50, c = 'red', label = "3")
plt.scatter(x[y_predict == 4,0],x[y_predict == 4,1],s = 50, c = 'pink', label = "4")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 100, c = "black", label = "centroids")
plt.xlabel("Answer -- >")
plt.ylabel("Question -- >")
plt.legend()
plt.show()

#centroids and clusters
```
