```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df_customerchurn = pd.read_csv('userchurn.csv')
```

```python
df_customerchurn
```

```python
df_customerchurn.drop(['customerID'],axis=1,inplace=True)
```

```python
df_customerchurn.isna()
```

```python
df_customerchurn.dropna(inplace=True)

#nan values dropped using dropna
```

```python
df_customerchurn
```

```python
df_customerchurn.duplicated()
```

```python
df_customerchurn.drop_duplicates(inplace=True)
#Duplicates removed
```

```python
df_customerchurn
```

```python
df_customerchurn.describe()
```

```python
df_customerchurn.corr()
```

```python
## y is dependent variable and X is independent variable.
# Load X Variables into a Pandas Dataframe with columns 
y=df_customerchurn['Churn']
X=df_customerchurn.drop(['Churn'],axis=1)
print(y.shape)
print(X.shape)
```

```python
print(X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)

#1a Split the data into test and  training and testing
```

```python
DC=DecisionTreeClassifier()
#1b build the model
DC=DC.fit(X_train,y_train)
y_pred=DC.predict(X_test)

#1c make Predictions
```

```python
print("Actual Values:")
print(y_test)
print("Predicted values:")
print(y_pred)
```

```python
d2=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
d2.head(30)

#actual and predicted dataframe
```

```python
print("accuracy",metrics.accuracy_score(y_test,y_pred))

#accuracy for knn:0.7518, acccuracy for SVC:0.7808, accuracy for RandomForest:0.7743, 
#accuracy for Decisiontreeclassifier:0.7213, accuracy for logisticregression:0.7837

#Logistic regression had the best accuracy of 0.7837
```

```python
cf=confusion_matrix(y_test,y_pred)
#confusion matrix
```

```python
print(cf)
#print confusion matrix#
```

```python
print(classification_report(y_test,y_pred))
#classification report#
#1d Evaluate model using appropriate evaluation metrics 

```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

clf = DecisionTreeClassifier(random_state=1)

params =  {
    'criterion' : ['gini', 'entropy'],
    'min_samples_leaf': [1, 2, 3],
    'max_depth': [1, 2, 3]
}
#
# Create gridsearch instance
#
grid = GridSearchCV(estimator=clf,param_grid=params,cv=10)
                    

grid.fit(X_train, y_train)
#
grid.best_score_, grid.best_params_

#grid best score: 0.7752
#grid best parameter: Criterion:gini, max_depth:3, Minimum sample leaf:1


#2a model built
#2b gridsearchcv used for parameter hypertuning 
```
