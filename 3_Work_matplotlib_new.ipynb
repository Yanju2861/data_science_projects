### 1.3 Object Oriented method of creating multiple plots (scattter, plot and/or stem) as a part of subplots()

```python
# Create just a figure and only one subplot
# Object oriented method of doing the same thing
fig , ax = plt.subplots( )
# ax.plot(x, y, marker='*')
ax.scatter( x, z,marker='^',c='black')
ax.set_title('My Scatter plot')
ax.set_xlabel('X value')
ax.set_ylabel('Z value')

### Saving the figure
fig.savefig('my_saved_oom.png', bbox_inches="tight")
# fig.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
```

# 3. USUAL PLOT PARAMETERS

```python
np.pi
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 40)
y = np.cos(x)
z = np.sin(x)
```

```python
## COLOR (color)
plt.plot(x, y, color='green')
plt.title('Cosine curve')
plt.show()
```

```python
## Line Width (lw)
plt.plot( x, y, color='red', linewidth=2)
# OR
# plt.plot(x, y, color='red', lw=5)

plt.title('Simple plot')
plt.show()
```

```python
## Line Transparency (alpha)
plt.plot(x, y, color='red', linewidth=2, alpha=0.8)
plt.title('Simple plot')
plt.show()
```

```python
## Line Style (linestyle)
plt.plot(x, y, color='red', linewidth=1, alpha=1, linestyle='dotted')
plt.title('Simple plot')
plt.show()
```

```python
## Line Markers (marker)
plt.plot(x, y, color='red', linewidth=1, alpha=1, marker='^')
plt.title('Simple plot')
plt.show()
```

```python
## Line Marker Size (marker size)
plt.plot(x, y, color='red', linewidth=1, alpha=1, marker='p', markersize=5)
plt.title('Simple plot')
plt.show()
```
