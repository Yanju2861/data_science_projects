```python
import pandas as pd 
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df_customerchurn= pd.read_csv('customerchurn.csv')
df_customerchurn

#Analyse the dataset and identify the task to be performed whether it is regression or
#classification
#The task is a classification problem according to the data
```

```python
## y is dependent variable and X is independent variable.
# Load X Variables into a Pandas Dataframe with columns 
y=df_customerchurn['Churn']
X=df_customerchurn.drop(['Churn'],axis=1)
print(y.shape)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#1a Split the data into training and testing dataset
#split 
```

```python
#feature Scaling  
#use standard scaler
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
X_train= st_x.fit_transform(X_train)    
X_test= st_x.transform(X_test) 
```

```python
#Fitting K-NN classifier to the training set  
from sklearn.svm import SVC
svm_classifier= SVC(kernel='rbf',degree=2,random_state=0)  # Hyper Parameter
svm_classifier.fit(X_train, y_train)  
##1b Build the model using training dataset
#2aModel building
```

```python
#Predicting the test set result  
y_pred= svm_classifier.predict(X_test)
print(y_pred)

#1c Make predictions using the built model on test dataset.
```

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred) 
```

```python
print(cm)
#print confusion matrix
```

```python
from sklearn import metrics
print("accuracy",metrics.accuracy_score(y_test,y_pred))

#Compare the accuracy of all 5 models and select the best one.
#SVC 0.780841,Logistic regression 0.783744,Knnclassifier 0.751814,Decisiontreeclassifier 0.722786,Randomforest Classifier0.775761
#Logistic regression has the highest accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
#1d Evaluate model using appropriate evaluation metrics 

#print classification report
```

```python
# Get Support Vector indices
support_vector_indices= svm_classifier.support_
print(support_vector_indices)
#print support vectors
```

```python
# Get number of support_vector per class
support_vector_per_class= svm_classifier.n_support_
print(support_vector_per_class)
```
